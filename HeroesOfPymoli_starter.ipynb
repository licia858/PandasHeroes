## Age Demographics

* Establish bins for ages


* Categorize the existing players using the age bins. Hint: use pd.cut()


* Calculate the numbers and percentages by age group


* Create a summary data frame to hold the results


* Optional: round the percentage column to two decimal points


* Display Age Demographics Table


```python
Age_bin = pd.DataFrame(uniqueplayers["Age"])                 #COMPLETED BELOW THE GROUPBY NEEDED 
bins = [0, 9, 14, 19, 24, 29, 34, 39, 100]  #bin
Age_range = ["< 10", "10-14", "15-19", "20-24", "25-29", "30-34", "35-39", "40+"] #range
Age_bin["Total Count"] = pd.cut(Age_bin["Age"], bins, labels= Age_range) #create var that will hold the cut values
Age_bin = Age_bin.groupby("Total Count") #group it in a list
Age_bin.count()#export will be used to merge the percent column
```

```python
#Find percent of age
uniqueplayerscount = uniqueplayercount["Age"].count()
Agecount = Age_bin["Total Count"].count()
Age_percent = pd.DataFrame(Agecount/uniqueplayerscount*100)
Age_percent = Age_percent.rename(columns = {"Total Count": "Age Precent"})
Age_count = pd.DataFrame(Agecount)
Age_count = Age_count.rename(columns = {"Total Count": "Age Count"})

Age_count = Age_count.merge(Age_percent, on="Total Count")
Age_count
```
